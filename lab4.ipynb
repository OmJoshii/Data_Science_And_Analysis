```python
# Lab 3: Chi-Squared Test of Independence in Python

# Step 1: Import required libraries
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
# Step 2: Create the observed frequency (contingency) table
# Example: Gender vs Product Preference
data = {
    "Like": [30,20],
    "Dislike": [10,40]
}
```

```python
observed = pd.DataFrame(data, index=["Male", "Female"])

print("Observed Frequency Table:")
print(observed)
```

```python
# Step 3: PErform Chi-Squared Test
chi2_stat, p_value, dof, expected = chi2_contingency(observed)
```

```python
# Step 4: Display results
print("\nChi-Square Statistic:",chi2_stat)
print("Degrees of Freedom",dof)
print("P-value:", p_value)

print("\nExpected Frequency Table:")
print(pd.DataFrame(expected,
                   index=observed.index,
                   columns=observed.columns))
```

```python
# Step 5: Hypothesis Decision
alpha = 0.05

if p_value < alpha:
    print("\nReject the Null Hypothesis")
    print("-> There is a significant association between the variables.")
else:
    print("\nFail to Reject the Null Hypothesis")
    print("-> There is no significant association between the variables.")
```
